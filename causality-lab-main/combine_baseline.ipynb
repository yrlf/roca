```python
import numpy as np
import pandas as pd
import os
```

```python
# get the filenames of all csv files in a directory
def get_csv_filenames(directory):

    files = os.listdir(directory)
    csv_files = []
    for file in files:
        if file.endswith('.csv'):
            csv_files.append(file)
    return csv_files


# read a list of csv files and combine them into a large data frame
def read_csv(file_list):
    df = pd.DataFrame()
    for file in file_list:
        # read csv file and concatenate to the large data frame
        df = pd.concat([df, pd.read_csv(file)], ignore_index=True)
        

    return df
```
